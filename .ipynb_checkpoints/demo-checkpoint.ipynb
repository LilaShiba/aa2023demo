```python
# Create H0/H1 from descriptive stats
# show rates without lables at first
def acceptanceRate(ethnicity,flag=False):
    
    accepted = df[ (df['accepted'] == 1) & (df['ethnicity'] == ethnicity)]
    notAccepted = df[(df['accepted'] == 0) & (df['ethnicity'] == ethnicity)]
    t = len(accepted)
    f = len(notAccepted)
    print(t/(t+f))
    return (accepted,notAccepted)
y,n = acceptanceRate('B')
y1,n1 = acceptanceRate('W')
y2,n2 = acceptanceRate('A')
y3,n3 = acceptanceRate('H')
```

```python
m = list(gpaCleaned['ethnicity'].unique())
n = range(len(m))
gpa = list(gpaCleaned['HS_GPA'])
y = list(gpaCleaned['ethnicity'])
accepted =list(gpaCleaned['accepted'])
colors = [i for i in n]
# accept vs. accepted tricky trick
gpaCleaned['cat'] = gpaCleaned['ethnicity']
gpaCleaned['cat'].replace(m,colors,inplace=True)
#gpaCleaned.head(5)
gpaCleaned.corr()
```

```python
class gpa:
    def __init__(self,mu,sigma,etho,col):
        self.mu = mu
        self.sigma = sigma
        self.etho = etho
        self.col = col
    
    def t_test(x,y):
        pass

res = []
def GPAByEthos(etho):
    d = { 
        0:'white',
        1:'black',
        2:'asian',
        3:'hispanic'
    }
    meow = gpaCleaned[ (gpaCleaned['cat'] == etho)]
    print(d[etho],':mu', round(meow['HS_GPA'].mean(),2),'sigma:', 
          round(meow['HS_GPA'].std(),2))
    obj = gpa(
        round(meow['HS_GPA'].mean(),2),
        round(meow['HS_GPA'].std(),2),
        d[etho],
        meow
    )
    
    obj.col['accepted'] = obj.col['accepted'].astype(float)
    return obj
    
for i in colors[0:4]:
    res.append(GPAByEthos(i))
```

```python
x = list(res[0].col['HS_GPA'])
y = list(res[1].col['HS_GPA'])
summary, results = sp.stats.ttest_ind(x,y)
```
